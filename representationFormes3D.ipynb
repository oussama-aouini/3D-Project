1. Enumérer et représenter toutes les classes vtk qui permettent la représentation d’objets
synthétiques (Superquadriques, ...).


```python
superquadricSource = vtk.vtkSuperquadricSource()
# superquadricSource = vtk.vtkConeSource()
# superquadricSource = vtk.vtkSphereSource()
# superquadricSource = vtk.vtkCubeSource()
# superquadricSource = vtk.vtkCylinderSource()
# superquadricSource = vtk.vtkPlatonicSolidSource()
# superquadricSource = vtk.vtkEllipticalButtonSource()


superquadricSource.SetThetaRoundness(1)
superquadricSource.SetPhiRoundness(2)

mapper = vtk.vtkPolyDataMapper()
mapper.SetInputConnection(superquadricSource.GetOutputPort())

actor = vtk.vtkActor()
actor.SetMapper(mapper)
colors = vtk.vtkNamedColors()
actor.GetProperty().SetColor(colors.GetColor3d("Cornsilk"))

renderer = vtk.vtkRenderer()
renderWindow = vtk.vtkRenderWindow()
renderWindow.SetWindowName("Sphere")
renderWindow.AddRenderer(renderer)
renderWindowInteractor = vtk.vtkRenderWindowInteractor()
renderWindowInteractor.SetRenderWindow(renderWindow)

renderer.AddActor(actor)
renderer.SetBackground(colors.GetColor3d("DarkGreen"))

renderWindow.Render()
renderWindowInteractor.Start()
```

3. Ecrire un programme qui permet de représenter des objets 3D au format .ply

```python
# SOURCE/READER
reader = vtk.vtkPLYReader()

reader.SetFileName('femure.ply')
reader.Update()
data = reader.GetOutput()

# MAPPER 
mapper = vtk.vtkPolyDataMapper()
mapper.SetInputData(data)

# ACTOR
actor = vtk.vtkActor()
actor.SetMapper(mapper)

# RENDERER
renderer = vtk.vtkRenderer()
renderer.AddActor(actor)

# RENDER-WINDOW
render_window = vtk.vtkRenderWindow()
render_window.AddRenderer(renderer)

# INTERACTOR
interactor = vtk.vtkRenderWindowInteractor()
interactor.SetRenderWindow(render_window)

interactor.Initialize()
render_window.Render()
interactor.Start()
```

6. Etudier le format à trois fichiers (fichier.points, fichier.links, fichier) joint à ce mail et écrire un programme qui permet de le représenter.

```python
'''Get the points of the object'''
#array of points
points = vtk.vtkPoints()

#read the points file
pointsFile = open('bunny.points','r') 

#get the number of points 
numberOfPoints = pointsFile.readline()

#get the number of axis 
numberOfCoordinates = pointsFile.readline()

for line in pointsFile:
    x, y, z, *others = line.split()
    x = float(x)
    y = float(y)
    z = float(z)
    points.InsertNextPoint(x,y,z)

pointsFile.close()
    
"""Make the triangles of the object"""

polys = vtk.vtkCellArray() #array of triangles

linksFile = open("bunny.links","r") #read the links file
numberOfLinks = int(linksFile.readline()) #get the number of points

polys.SetNumberOfCells(numberOfLinks)

for line in linksFile:
    c, id1, id2, id3 = line.split()
    id1 = int(id1)
    id2 = int(id2)
    id3 = int(id3)
    triangle = vtk.vtkTriangle()
    triangle.GetPointIds().SetId(0,id1)
    triangle.GetPointIds().SetId(1,id2)
    triangle.GetPointIds().SetId(2,id3)
    polys.InsertNextCell(triangle)
    
linksFile.close()

#create the poly data
polyData = vtk.vtkPolyData()
polyData.SetPoints(points)
polyData.SetPolys(polys)

#mapper
mapper = vtk.vtkPolyDataMapper()
mapper.SetInputData(polyData)

#actor
actor = vtk.vtkActor()
actor.SetMapper(mapper)

renderer = vtk.vtkRenderer()
renderWindow = vtk.vtkRenderWindow()
renderWindow.SetWindowName("TP1")
renderWindow.AddRenderer(renderer)

#an interactor
iren = vtk.vtkRenderWindowInteractor()
iren.SetRenderWindow(renderWindow)
renderer.AddActor(actor)
renderer.SetBackground(0,1,0)
renderWindow.Render()
iren.Start()
```
